```python
from pyspark.sql import SparkSession

# SparkSession
spark = SparkSession.builder.appName("LoanDefaultPrediction").getOrCreate()
```

```python
df = spark.read.csv("/content/Loan_Default.csv", header=True, inferSchema=True)
df.show(5)  #data frame check
df.printSchema()  # Data Schema check
```

```python
print(df)
```

```python
df.select([count(when(col(c).isNull(), c)).alias(c) for c in df.columns]).show()
```

```python
# Null value check
df.select([count(when(col(c).isNull(), c)).alias(c) for c in df.columns]).show()

# Null value drop
df = df.dropna()


```

**Add New Column**

```python
from pyspark.sql.functions import when

df = df.withColumn("loan_status", when(df["Status"] == 1, "Default").otherwise("Non-Default"))
df.show(5)
```

***Data Type Change ***

```python
df.columns
```

```python
from pyspark.sql.functions import col

# Cast the 'loan_amount' column to double
df = df.withColumn("loan_amount", col("loan_amount").cast("double"))

# Cast the 'rate_of_interest' column to double (previously 'interest_rate')
df = df.withColumn("rate_of_interest", col("rate_of_interest").cast("double"))

df.printSchema()
```

```python
from pyspark.sql.functions import col

# loan_amount column double type cust
df = df.withColumn("loan_amount", col("loan_amount").cast("double"))

# rate_of_interest column double type cust
df = df.withColumn("rate_of_interest", col("rate_of_interest").cast("double"))

# Schema show
df.printSchema()
```

**string indexer**

```python
from pyspark.ml.feature import StringIndexer

indexer = StringIndexer(inputCol="loan_status", outputCol="loan_status_index")
df = indexer.fit(df).transform(df)
df.show(5)
```

**Feature Vectorization **

```python
from pyspark.ml.feature import VectorAssembler

# Feature Column Update
feature_cols = ['loan_amount', 'rate_of_interest', 'loan_status_index']

# VectorAssembler Create
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")

# Transformation Apply
df = assembler.transform(df)

# Feture  Column show
df.select("features").show(5, truncate=False)
```

**data split **

```python
train_data, test_data = df.randomSplit([0.75, 0.25], seed=42)
```

**Model Train **

```python
from pyspark.ml.classification import LogisticRegression

lr = LogisticRegression(featuresCol="features", labelCol="loan_status_index")
lr_model = lr.fit(train_data)
```

**Model Test and Evaluation **

```python
predictions = lr_model.transform(test_data)
predictions.select("loan_status_index", "prediction").show(10)

from pyspark.ml.evaluation import MulticlassClassificationEvaluator

evaluator = MulticlassClassificationEvaluator(labelCol="loan_status_index", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print(f"Model Accuracy: {accuracy:.2f}")
```

**Prediction & Deployment**

**Model Save **

```python
lr_model.save("/content/logistic_regression_model")
```

Prediction by new data

```python
new_data = spark.createDataFrame([
    (20000, 7.5, 1.0),
    (5000, 5.2, 0.0)
], ["loan_amount", "rate_of_interest", "loan_status_index"])

new_data = assembler.transform(new_data)
predictions = lr_model.transform(new_data)
predictions.select("features", "prediction").show()
```
